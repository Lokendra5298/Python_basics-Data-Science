## Q.5. WAP to implement a class called "Bank_Account" representing a person's bank account.
The class should have the following attributes: account_holder_name (str), account_number(int), address (str) and balance (float).
The class should have methods to implement the following:
    deposit - Deposits a given amount into the account
    withdraw - Withdraws a given amount from the account
    check_balance - To get the current balance
    update_details - To update the name and address from the user and displays a message indicating successful update
    display_details - To display the details of the account.

```python
class Bank_Account:
    def __init__(self, account_holder_name='', account_number=0, address='', balance=0.0):
        self.Name = account_holder_name
        self.Account_no = account_number
        self.Address = address
        self.Balance = balance

    def display_details(self):
        print('Customer Name:', self.Name, ", Account No:", self.Account_no, ", Address:", self.Address, ', Balance:', self.Balance)

    def deposit(self, deposit_amt):
        self.Balance = self.Balance + deposit_amt
        print('Total amount after deposit is', self.Balance)

    def withdraw(self, withdraw_amt):
        if self.Balance >= withdraw_amt:
            self.Balance = self.Balance - withdraw_amt
            print('Total amount after withdrawal is', self.Balance)
        else:
            print('Insufficient funds! Withdrawal not allowed.')

    def update_details(self, new_name, new_address):
        self.Name = new_name
        self.Address = new_address
        print('Account details updated successfully.')

    def check_balance(self):
        print('Current balance is', self.Balance)


# Example Usage:
a1 = Bank_Account('Lokendra', 681461481648, 'Hamirpur', 6514.32)
a1.display_details()

a1.deposit(500)
a1.display_details()

a1.withdraw(200)
a1.display_details()

a1.update_details('Ram', 'Lucknow')
a1.display_details()

a1.check_balance()
```

## Q.6.  Define a Matrix class of dimensions m X n (the values for m and n can be taken as input). Demonstrate matrix addition, subtraction, multiplication, element-by-element multiplication, scalar multiplication (use map here). Use operator overloading wherever possible. (Hint: In the constructor, use 'random' and create list of list using list comprehension. In the arguments of constructor, send the number of rows and columns)
Ensure that your implementation follows best practices for class design and encapsulation in Python. Comment your code to explain the functionality of each method.

```python
import random

class Matrix_class:
    def __init__(self, m=0, n=0):
        self.N_rows = m
        self.N_columns = n
        self.Matrix = [[random.randint(1, 10) for _ in range(n)] for _ in range(m)]

    # String representation of the matrix.
    def __str__(self):
        return '\n'.join([' '.join(map(str, row)) for row in self.Matrix])

    # Overloaded '+' operator for matrix addition.
    def __add__(self, other):
        if self.N_rows == other.N_rows and self.N_columns == other.N_columns:
            result = Matrix_class(self.N_rows, self.N_columns)
            result.Matrix = [[self.Matrix[i][j] + other.Matrix[i][j] for j in range(self.N_columns)] for i in range(self.N_rows)]
            return result
        else:
            print("Matrices must have the same dimensions for addition")
            return None
            
    def __sub__(self, other):
        if self.N_rows == other.N_rows and self.N_columns == other.N_columns:
            result = Matrix_class(self.N_rows, self.N_columns)
            result.Matrix = [[self.Matrix[i][j] - other.Matrix[i][j] for j in range(self.N_columns)] for i in range(self.N_rows)]
            return result
        else:
            print("Matrices must have the same dimensions for subtraction")
            return None

    # Overloaded the '*' operator for matrix multiplication.
    def __mul__(self, other):
        if isinstance(other, Matrix_class):
            if self.N_columns == other.N_rows:
                result = Matrix_class(self.N_rows, other.N_columns)
                for i in range(self.N_rows):
                    for j in range(other.N_columns):
                        result.Matrix[i][j] = sum(self.Matrix[i][k] * other.Matrix[k][j] for k in range(self.N_columns))
                return result
            else:
                print("Matrix multiplication is not possible")
                return None
        else:
            # Scalar multiplication
            result = Matrix_class(self.N_rows, self.N_columns)
            result.Matrix = [[element * other for element in row] for row in self.Matrix]
            return result
            
    def elementwise_multiply(self, other):
        if self.N_rows == other.N_rows and self.N_columns == other.N_columns:
            result = Matrix_class(self.N_rows, self.N_columns)
            result.Matrix = [[self.Matrix[i][j] * other.Matrix[i][j] for j in range(self.N_columns)] for i in range(self.N_rows)]
            return result
        else:
            print("Matrices must have the same dimensions for element-wise multiplication")
            return None


# Example usage:
m1 = Matrix_class(2, 2)
m2 = Matrix_class(2, 2)  # Adjusted the dimensions for demonstration purposes

print("Matrix 1:")
print(m1)

print("\nMatrix 2:")
print(m2)

result_addition = m1 + m2
print("\nMatrix Addition:")
print(result_addition)

result_sub = m1 - m2
print("\nMatrix Subtraction:")
print(result_sub)

result_mul = m1 * m2
print("\nMatrix Multiplication:")
print(result_mul)

elementwise_mul = m1.elementwise_multiply(m2)  # Element-wise multiplication
print("\nElement-by-element Multiplication:")
print(elementwise_mul)

scalar_mul = m1 * 2  # Scalar multiplication
print("\nScalar Multiplication:")
print(scalar_mul)
```

## Q.7. Create a Python class named Time that represents a moment of time. The class should have attributes hour, minute, and second. Include the following features:
    A constructor that initializes hour, minute, and second, with validation to ensure each attribute is within its correct range (hours: 0-23, minutes: 0-59, seconds: 0-59).
    A __str__() method that returns the time in a format hh:mm:ss.
    Methods set_time(hour, minute, second) and get_time() to update and access the time, respectively.
    An add_seconds(seconds) method that adds a given number of seconds to the current time object, adjusting the hour, minute, and second attributes accordingly.

```python
class Time:
    def __init__(self, hour=0, minute=0, second=0):
        # Ensure hour, minute, and second are within valid ranges
        if 0 <= hour <= 23 and 0 <= minute <= 59 and 0 <= second <= 59:
            self.hour = hour
            self.minute = minute
            self.second = second
        else:
            print("Invalid time, Hour and minute values.")

    def __str__(self):
        return f"hh:mm:ss = {self.hour:02d}:{self.minute:02d}:{self.second:02d}"
        
    def set_time(self, hour, minute, second):
        if 0 <= hour <= 23 and 0 <= minute <= 59 and 0 <= second <= 59:
            self.hour = hour
            self.minute = minute
            self.second = second
        else:
            print("Invalid time, Hour and minute values.")
            
    def get_time(self):
        return self.hour, self.minute, self.second
        
    def add_seconds(self, seconds):
        # Add the given number of seconds to the current time
        total_seconds = self.hour * 3600 + self.minute * 60 + self.second + seconds
        self.hour = (total_seconds // 3600) % 24
        self.minute = (total_seconds // 60) % 60
        self.second = total_seconds % 60

# Example usage
time_obj = Time(12, 34, 56)
print(str(time_obj)) 

time_obj.set_time(15, 23, 42)
print('Updated time :')
print(str(time_obj)) 

# Access the time using get_time()
current_time = time_obj.get_time()
print('current_time: ', current_time)

# Add 115 seconds to the time
time_obj.add_seconds(115)
print('Time after adding 115 seconds = ', time_obj.get_time())
```

## Q. 8.  Create a class named Geometry that provides static methods for various geometric calculations, such as area and perimeter, for different shapes (circle, rectangle, square). Include:
Static methods like circle_area(radius), rectangle_area(length, width), and square_area(side).
Static methods for perimeter calculations for the above shapes.
Ensure that methods check for valid inputs (e.g., positive numbers).

```python
import math

class Geometry:
    # Calculate the area of a circle.
    @staticmethod
    def circle_area(radius):
        if radius > 0:
            return math.pi * radius**2
        else:
            print("Radius must be a positive number")

    # Calculate the perimeter of a circle.
    @staticmethod
    def circle_perimeter(radius):
        if radius > 0:
            return 2*math.pi * radius
        else:
            print("Radius must be a positive number")

    # Calculate the area of a rectangle.
    @staticmethod
    def rectangle_area(length, width):
        if length >= 0 and width >= 0:
            return length * width
        else:
            print("Length and width must be positive numbers")

    # Calculate the perimeter of a rectangle.
    @staticmethod
    def rectangle_perimeter(length, width):
        if length >= 0 and width >= 0:
            return 2 * (length + width)
        else:
            print("Length and width must be positive numbers")

    # Calculate the area of a square.
    @staticmethod
    def square_area(side):
        if side >= 0:
            return side**2
        else:
            print("Side length must be a positive number")

    # Calculate the perimeter of a square.
    @staticmethod
    def square_perimeter(side):
        if side >= 0:
            return 4 * side
        else:
            print("Side length must be a positive number")


geometry_instance = Geometry()
circle_radius = 5
print(f"Circle Area: {geometry_instance.circle_area(circle_radius)}")
print(f"Circle Perimeter: {geometry_instance.circle_perimeter(circle_radius)}")

rectangle_length = 4
rectangle_width = 6
print(f"Rectangle Area: {geometry_instance.rectangle_area(rectangle_length, rectangle_width)}")
print(f"Rectangle Perimeter: {geometry_instance.rectangle_perimeter(rectangle_length, rectangle_width)}")

square_side = 3
print(f"Square Area: {geometry_instance.square_area(square_side)}")
print(f"Square Perimeter: {geometry_instance.square_perimeter(square_side)}")
```
