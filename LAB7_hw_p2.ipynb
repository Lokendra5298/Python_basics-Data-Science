## Q.5. WAP to plot a 3-d graph of the helical wave signal using the scatter method and normal line method. Plot them separately and specify legend.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to generate helical wave signal
def helical_wave(t):
    x = np.sin(t)
    y = np.cos(t)
    z = t
    return x, y, z

t = np.linspace(0, 10*np.pi, 1000)
x_values, y_values, z_values = helical_wave(t)

# Plot using scatter method
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_values, y_values, z_values, label='Scatter Method')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.legend()

plt.title('3D Scatter Plot of Helical Wave Signal')
plt.show()
```

```python
# Plot using normal line method
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_values, y_values, z_values, label='Normal Line Method')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.legend()

# Show the plot
plt.title('3D Line Plot of Helical Wave Signal')
plt.show()
```

## Q.8.Take a N X M integer array matrix with space separated elements ( N = rows and M  = columns). Your task is to print the transpose and flatten results using numpy.

```python
import numpy as np

# Input matrix as a string (replace this with your actual data)
matrix_str = "11 2 6; -5 0 6; 4 -8 12"

# Convert the string matrix to a 2D NumPy array
matrix = np.array([list(map(int, row.split())) for row in matrix_str.split(';')])

# Print the original matrix
print("Original Matrix:")
print(matrix)
print('\n')
# Print the transpose of the matrix
transpose_matrix = np.transpose(matrix)
print("Transpose Matrix:")
print(transpose_matrix)
print('\n')
# Print the flattened version of the matrix
flattened_matrix = matrix.flatten()
print("Flattened Matrix:")
print(flattened_matrix)
```

## Q.9. WAP to capitalize a column of names in a Pandas Dataframe.

Eg : Input : {'Name': ['john', 'bODAY', 'aNa', 'Peter', 'nicky'], 'Education': ['masters', 'graduate', 'graduate', 'Masters', 'Graduate'], 'Age': [27, 23, 21, 23, 24]}

```python

Input= {'Name': ['john', 'bODAY', 'aNa', 'Peter', 'nicky'], 
        'Education': ['masters', 'graduate', 'graduate', 'Masters', 'Graduate'], 
        'Age': [27, 23, 21, 23, 24]
       }
df1 = pd.DataFrame(Input)

# Applying the function to the 'Name' column
df1['Name'] = df1['Name'].apply(lambda x: x.capitalize())

# Converting DataFrame to the dictionary
output_dict = df1.to_dict(orient='list')

print(output_dict)
```

## Q.10. Use the central difference method to find the first and second order derivatives of the function. Use the following function for testing the result. And also verify the result manually (Write on paper and upload jpg). Refer to section 2.5.1 of “Optimization for Engineering Design: Algorithms and Examples” by KALYANMOY DEB, 2nd edition

f(x) = 3x**2 + 2x

```python
def f(x):
    return 3*(x**2) + 2*x

def first_der(x, h):
    der_f = (f(x+h) - f(x-h))/(2*h)
    return der_f

def sec_der(x, h):
    der_f = (f(x+h) - 2*f(x) + f(x-h))/(h**2)
    return der_f

result_first_der = first_der(1, 0.0001)
result_sec_der = sec_der(1, 0.0001)

print("First derivative:", result_first_der)
print("Second derivative:", result_sec_der)
```
