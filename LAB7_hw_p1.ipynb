## Q.1. Write a program that takes coefficients A, B, C, D, and E as inputs representing a 4th degree polynomial in the form Ax^4 + Bx^3 + Cx^2 + Dx + E. Calculate the values of this polynomial for x in the range from -100 to 100, with constant discrete intervals.

Store the resulting x and y values as a NumPy array, where x represents the input values, and y represents the corresponding output values of the polynomial. Finally, use Matplotlib to plot the graph using the generated NumPy array.

```python
import numpy as np
import matplotlib.pyplot as plt

def polynomial(A, B, C, D, E, x):
    f_x = A*x**4 + B*x**3 + C*x**2 + D*x + E
    return f_x

A = float(input('Enter the value of coefficient A: '))
B = float(input('Enter the value of coefficient B: '))
C = float(input('Enter the value of coefficient C: '))
D = float(input('Enter the value of coefficient D: '))
E = float(input('Enter the value of coefficient E: '))

x_values = np.linspace(-100, 100, 50)
y_values = polynomial(A, B, C, D, E, x_values)

print(x_values)
print(y_values)

# plot the graph using the generated NumPy array
plt.plot(x_values, y_values, 'o--')
plt.title('Polynomial Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

## Q.2. Suppose you have a dictionary containing information about monthly sales for different products over a period of time. The dictionary has the following structure.

sales_data = {

    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],

    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],

    'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]

}

Write a Python script to convert this dictionary into a pandas DataFrame, calculate the total sales for each product over the entire period, and then create a bar plot using matplotlib to visualize the total sales for each product.

```python
import pandas as pd
import matplotlib.pyplot as plt

#Dictionary
sales_data = { 'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],
              'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],
              'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]
             }

df=pd.DataFrame(sales_data)
print(df)
total_sales_A = df[df['Product'] == 'A']['Sales'].sum()
total_sales_B = df[df['Product'] == 'B']['Sales'].sum()
total_sales_C = df[df['Product'] == 'C']['Sales'].sum()

sales_detail = { 'Products': ['A', 'B', 'C'],
                'Total_sales' : [total_sales_A, total_sales_B, total_sales_C] 
               }
df_total = pd.DataFrame(sales_detail)
print(df_total)

df_total.plot.bar(x='Products', y='Total_sales', rot=0, color='skyblue', legend=False)
```

## Q.3 (b) Generate surface plots for these multi-variable functions over the range 

x=[−10,10] and y=[−10,10] , ensuring to add a title and labels for all axes:

(1) z = cos(sqrt(x^2+y^2)

(2) z = e^(-(x^2+y^2))

(3) z =  log(x^2+y^2) where x^2+y^2>0

```python
# z = cos(sqrt(x^2+y^2)

u = np.linspace(-10, 10, 100)
v = np.linspace(-10, 10, 100)

# The following will print a 3D surface
X,Y=np.meshgrid(u,v) 
Z = np.cos(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```

```python
# z = e^(-(x^2+y^2))

Z1 = np.exp(-(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z1)
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```

```python
# z = log(x^2+y^2) where x^2+y^2>0

u1 = np.linspace(-10, 10, 100)
v1 = np.linspace(-10, 10, 100)

X,Y=np.meshgrid(u1,v1) 
Z2 = np.log(X**2 + Y**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z2)
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```

## Q.4. For the function J(w) = w^2 + (54/w), implement the bracketing method (choose your own a, b, n).


```python
def bracketing(J, a, b, n):
    
    delta_w = (b-a)/n
    w1 = a
    w2 = a + delta_w
    w3 = a + 2*delta_w

    # Step -2
    flag = 0
    while flag !=1:
        if J(w1) >= J(w2) <= J(w3):
            print('Minimum lies between', w1, 'and', w3)
            flag =1
        else:
            # Update points for the next iteration
            w1 = w2
            w2 = w3
            w3 = w3 + delta_w

        if w3 > b:
            print('No minimum exists between a, b')
            flag = 1

# Example usage:
def J(w):
    return w**2 + (54/w)

# Set the interval [a, b] and the number of intermediate points (n)
a = 1
b = 6
n = 100

bracketing(J, a, b, n)
```
